```python
import pandas as pd
import plotly.express as px
```

```python
f = '../data/trad.csv'
data = pd.read_csv(f, index_col=0, parse_dates=True)
data
```

```python
fig = px.scatter(data, x=data.index, y=['temp','GLOBO'])
fig.show()
```

```python
fig = px.scatter(data, x=data.temp, y=data.GLOBO)
fig.show()
```

```python
data['error'] = data['temp'] - data['GLOBO']
```

```python
ME = data['error'].mean()
MAE = data['error'].abs().mean()
```

```python
print("ME:", ME)
print("(MAE:", MAE)
```
