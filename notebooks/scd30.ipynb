```python
import pandas as pd
import plotly.express as px
```

```python
f = '../data/scd30.parquet'
data = pd.read_parquet(f)
data
```

```python
fig = px.scatter(data, x=data.index, y=['HR', 'T', 'CO2', 'HR_F', 'T_F', 'CO2_F'])
fig.update_layout(hovermode="x unified")
fig.show()
```

## T

```python
fig = px.scatter(data, x=data["T"], y=data["T_F"])
fig.show()
```

```python
data['t_error'] = data['T'] - data['T_F']
T_ME = data['t_error'].mean()
T_MAE = data['t_error'].abs().mean()
print("ME:", T_ME)
print("MAE:", T_MAE)
```

## HR

```python
fig = px.scatter(data, x=data["HR"], y=data["HR_F"])
fig.show()
```

```python
data['hr_error'] = data['HR'] - data['HR_F']
HR_ME = data['hr_error'].mean()
HR_MAE = data['hr_error'].abs().mean()
print("ME:", HR_ME)
print("MAE:", HR_MAE)
```

## CO2

```python
fig = px.scatter(data, x=data["CO2"], y=data["CO2_F"])
fig.show()
```

```python
data['co2_error'] = data['CO2'] - data['CO2_F']
CO2_ME = data['co2_error'].mean()
CO2_MAE = data['co2_error'].abs().mean()
print("ME:", CO2_ME)
print("MAE:", CO2_MAE)
```
